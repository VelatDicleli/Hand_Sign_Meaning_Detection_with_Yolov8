```python
from ultralytics import YOLO
import cv2
import math



mdll = YOLO('handbest.pt')

names =["Iyi sanslar","Onaylandi","Onaylanmadi","Seni seviyorum","Tamam"]
```

```python
cap = cv2.VideoCapture(0)



while True:
    ret, frame = cap.read()
    results= mdll(frame,stream=True)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

            conf = math.ceil((box.conf[0] * 100)) / 100
            cls = int(box.cls[0])
            currentClass = names[cls]
            
        
            cv2.rectangle(frame, (x1, y1), (x2, y2), (150, 255, 0), 3)
            cv2.putText(frame, currentClass, (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 120, 255), 2)
           

    cv2.imshow("screen", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
